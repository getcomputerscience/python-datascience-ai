##A pivot table is a similar operation that seen in spreadsheets and other programs that operate on tabular data. 

#pivot table:-

##It takes simple column-wise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('profitdata.xlsx')
df
```

```python
pd.pivot_table(df,index=["Region","Software Engineer"])
```

```python
pd.pivot_table(df,index = ["Region","Manager"], values = ["Profit"],aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=["Name of Projects"], values="Number of Projects", aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=["Region", "Name of Projects"], 
               values="Number of Projects", aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=["Manager"],values=["Profit"],aggfunc=[np.mean,len])
```

```python
table = pd.pivot_table(df,index=["Region","Manager"],
                       values=["Number of Projects","Profit"],
                       aggfunc=[np.sum],fill_value=0,margins=True)
table
```

```python
table = pd.pivot_table(df,index=["Region","Manager","Software Engineer"], values="Profit")
table.query('Manager == ["Sita"]')
```

```python
table = pd.pivot_table(df, index='Number of Projects', values='Profit', aggfunc=np.max)
table
```

```python
table = pd.pivot_table(df, index='Number of Projects', values='Profit', aggfunc=np.min)
table
```

```python
table = pd.pivot_table(df, index='Manager', values='Number of Hours', 
                       aggfunc=[np.max, np.min])
table
```
