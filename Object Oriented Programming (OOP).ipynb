# Classes

Trying to access an attribute of an instance that isn't defined causes an AttributeError. This also applies when you call an undefined method.

```python
class Rectangle: 
  def __init__(self, width, height):
    self.width = width
    self.height = height

rect = Rectangle(7, 8)
print(rect.color)
```

# __init__

The __init__ method is the most important method in a class.

This is called when an instance (object) of the class is created, using the class name as a function.

```python
class list:
  def __init__(self, integer, floatt):
    self.interger = integer
    self.floatt = floatt

a = list(4, 5.6)
print(a.interger)

#the __init__ method takes two arguments and 
#assigns them to the object's attributes. 
#The __init__ method is called the class constructor.
```

# Methods

Classes can have other methods defined to add functionality to them.
Remember, that all methods must have self as their first parameter.
These methods are accessed using the same dot syntax as attributes.

```python
class list:
  interger=4  
  def __init__(self, integer, floatt):
    self.interger = integer
    self.floatt = floatt


a = list(2, 5.6)
print(list.interger)
```

# Inheritance

Inheritance provides a way to share functionality between classes.
Imagine several classes, Horse, Dog, Rabbit and so on. 

Although they may differ in some ways (only Dog might have the method bark), they are likely to be similar in others (all having the attributes color and name).

This similarity can be expressed by making them all inherit from a superclass Animal, which contains the shared functionality.

To inherit a class from another class, put the superclass name in parentheses after the class name.

```python
class Animal: 
  def __init__(self, name, color):
    self.name = name
    self.color = color

class horse(Animal):
  def purr(self):
    print("tabby...")
        
class Dog(Animal):
  def bark(self):
    print("tommy!")

jally = Dog("Jally", "black")
print(jally.color)
jally.bark()
```

# A class that inherits from another class is called a subclass.

A class that is inherited from is called a superclass.
If a class inherits from another with the same attributes or methods, it overrides them.

```python
class horse: 
  def __init__(self, name, color):
    self.name = name
    self.color = color

  def meow(self):
    print("meow...")

class cat(horse):
  def meow(self):
    print("horse")
        
parsian = cat("Max", "black")
parsian.meow()


#In the example above, Horse is the superclass, Cat is the subclass.
```
