Histogram

```python
#importing the Library
import matplotlib.pyplot as plt

import numpy as np

import random
```

```python
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
#hist
plt.hist(ml_students_age)

plt.title("Ml students age")

plt.xlabel("Students age category")

plt.ylabel("Number of students age")

plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.hist(ml_students_age,bins)

plt.title("Ml students age")

plt.xlabel("Students age category")

plt.ylabel("Number of students age")
plt.show()
```

```python
#Applying Width
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8)
plt.title("Ml students age")
plt.xlabel("Students age category")
plt.ylabel("Number of students age")
plt.show()
```

```python
#histtype
#Applying Width
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="step")
plt.title("Ml students age")
plt.xlabel("Students age category")
plt.ylabel("Number of students age")
plt.show()
```

```python
#histtype
#Applying histtype
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar")
plt.title("Ml students age")
plt.xlabel("Students age category")
plt.ylabel("Number of students age")
plt.show()
```

```python
#Orientation
#histtype
#Applying Width
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal")
plt.title("Ml students age")
plt.xlabel("Students age category")
plt.ylabel("Number of students age")
plt.show()
```

```python
#color
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal",color="r")
plt.title("Ml students age")
plt.xlabel("Students age category")
plt.ylabel("Number of students age")
plt.show()
```

```python
#Label
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal",color="r",label="ML Students")
plt.title("Ml students age")
plt.xlabel("Students age category")
plt.ylabel("Number of students age")
plt.legend()
plt.show()
```

```python
#color
bins=[15,20,25,30,35,40,45]

plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",orientation="vertical",color="m",label="ML Student")
plt.hist(py_students_age,bins,rwidth=0.8,histtype="bar",orientation="vertical",color="y",label="PY Student")

plt.title("Ml students age")
plt.xlabel("Students age category")
plt.ylabel("Number of students age")
plt.legend()
plt.show()
```

Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["Python","R","AI","ML","DS"]
class1_Students=[30,10,20,25,10]
class2_Students=[40,5,20,20,10]
class3_Students=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_Students)
```

```python
plt.barh(classes,class1_Students)
```

```python
plt.bar(classes,class1_Students,width=0.3,align="edge",label="class1_students",color="y")

plt.title("Class1 Students")
plt.xlabel("Courses")
plt.ylabel("Students")
plt.legend()
plt.show()
```

```python
plt.bar(classes,class1_Students,width=0.3,align="edge",color="y",edgecolor="r",linewidth=3)

plt.title("Class1 Students")
plt.xlabel("Courses")
plt.ylabel("Students")
plt.show()
```

```python
#xticks=(ticks=None,labels=None)
plt.figure(figsize=(15,9))
class_index=np.arange(len(classes))
width=0.2

plt.bar(class_index,class1_Students,width,color="b",label="class1 Students")
plt.bar(class_index+width,class2_Students,width,color="g",label="class2 Students")
plt.bar(class_index+width+width,class3_Students,width,color="y",label="class3 Students")

plt.xticks(class_index+width,classes,rotation=20)
plt.title("Bar Chart of CLASSES")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of students",fontsize=15)
plt.show()
```

```python
#yticks=(ticks=None,labels=None)
plt.figure(figsize=(15,9))
class_index=np.arange(len(classes))
width=0.2

plt.barh(class_index,class1_Students,width,color="b",label="class1 Students")
plt.barh(class_index+width,class2_Students,width,color="g",label="class2 Students")
plt.barh(class_index+width+width,class3_Students,width,color="y",label="class3 Students")

plt.yticks(class_index+width,classes,rotation=20)
plt.title("Bar Chart of CLASSES")
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of students",fontsize=15)
plt.show()
```
